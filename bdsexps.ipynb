```python
def caesar_encrypt(plain_text, shift):
    encrypted_text = ""

    for char in plain_text:
        if char.isalpha():
            is_upper = char.isupper()
            char = char.lower()
            shifted_char = chr(((ord(char) - ord('a') + shift) % 26) + ord('a'))
            if is_upper:
                shifted_char = shifted_char.upper()
            encrypted_text += shifted_char
        else:
            encrypted_text += char

    return encrypted_text

def caesar_decrypt(encrypted_text, shift):
    decrypted_text = ""

    for char in encrypted_text:
        if char.isalpha():
            is_upper = char.isupper()
            char = char.lower()
            shifted_char = chr(((ord(char) - ord('a') - shift) % 26) + ord('a'))
            if is_upper:
                shifted_char = shifted_char.upper()
            decrypted_text += shifted_char
        else:
            decrypted_text += char

    return decrypted_text

def main():
    message = input("Enter the message: ")
    shift = int(input("Enter the shift value: "))

    encrypted_message = caesar_encrypt(message, shift)
    decrypted_message = caesar_decrypt(encrypted_message, shift)
    print("\nEncrypted Message:")
    print(encrypted_message)
    print("\nDecrypted Message:")
    print(decrypted_message)
if __name__ == "__main__":
    main()
```

```python
def create_substitution_table(key):
    # Create a substitution table based on the given key
    alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    substitution_table = {}

    for i in range(len(alphabet)):
        substitution_table[alphabet[i]] = key[i]

    return substitution_table

def substitution_encrypt(plain_text, substitution_table):
    encrypted_text = ""
    for char in plain_text:
        if char.isalpha():
            is_upper = char.isupper()
            char = char.upper()
            if char in substitution_table:
                encrypted_char = substitution_table[char]
                if is_upper:
                    encrypted_char = encrypted_char.upper()
                encrypted_text += encrypted_char
            else:
                encrypted_text += char
        else:
            encrypted_text += char

    return encrypted_text

def substitution_decrypt(encrypted_text, substitution_table):
    decryption_table = {v: k for k, v in substitution_table.items()}
    decrypted_text = ""
    for char in encrypted_text:
        if char.isalpha():
            is_upper = char.isupper()
            char = char.upper()
            if char in decryption_table:
                decrypted_char = decryption_table[char]
                if is_upper:
                    decrypted_char = decrypted_char.upper()
                decrypted_text += decrypted_char
            else:
                decrypted_text += char
        else:
            decrypted_text += char

    return decrypted_text

def main():
    key = input("Enter the substitution key (26 unique letters): ").upper()
    if len(key) != 26 or not key.isalpha() or len(set(key)) != 26:
        print("Invalid key. Please enter 26 unique letters of the alphabet.")
        return

    substitution_table = create_substitution_table(key)

    message = input("Enter the message: ")

    encrypted_message = substitution_encrypt(message, substitution_table)
    decrypted_message = substitution_decrypt(encrypted_message, substitution_table)

    print("\nEncrypted Message:")
    print(encrypted_message)

    print("\nDecrypted Message:")
    print(decrypted_message)

if __name__ == "__main__":
    main()
```

```python
import string

def calculate_letter_frequencies(text):
    # Count the frequency of each letter in the text (case-insensitive)
    frequencies = {char: 0 for char in string.ascii_lowercase}
    total_letters = 0

    for char in text:
        if char.isalpha():
            char = char.lower()
            frequencies[char] += 1
            total_letters += 1

    # Calculate the relative frequencies
    relative_frequencies = {char: freq / total_letters for char, freq in frequencies.items()}

    # Sort the frequencies in descending order
    sorted_frequencies = sorted(relative_frequencies.items(), key=lambda x: x[1], reverse=True)

    return sorted_frequencies

def decrypt_with_substitution(text, substitution_map):
    # Decrypt the text using the provided substitution map
    decrypted_text = ""
    for char in text:
        if char.isalpha():
            char = char.lower()
            decrypted_char = substitution_map.get(char, char)
            if char.isupper():
                decrypted_char = decrypted_char.upper()
            decrypted_text += decrypted_char
        else:
            decrypted_text += char

    return decrypted_text

def main():
    ciphertext = input("Enter the ciphertext: ")
    letter_frequencies = calculate_letter_frequencies(ciphertext)

    # Define a frequency table for the English language (you can adjust this based on your needs)
    english_letter_frequencies = "etaoinshrdlcumwfgypbvkjxqz"

    # Create a substitution map based on the frequency analysis
    substitution_map = {}
    for i in range(len(letter_frequencies)):
        cipher_char = letter_frequencies[i][0]
        most_likely_plain_char = english_letter_frequencies[i]
        substitution_map[cipher_char] = most_likely_plain_char

    decrypted_text = decrypt_with_substitution(ciphertext, substitution_map)

    print("\nFrequency Analysis:")
    for cipher_char, freq in letter_frequencies:
        print(f"{cipher_char}: {freq:.4f}")

    print("\nSubstitution Map:")
    for cipher_char, plain_char in substitution_map.items():
        print(f"{cipher_char} -> {plain_char}")

    print("\nDecrypted Text:")
    print(decrypted_text)

if __name__ == "__main__":
    main()
```

```python
def initialize_rc4(key):
    key_length = len(key)
    S = list(range(256))
    j = 0

    for i in range(256):
        j = (j + S[i] + key[i % key_length]) % 256
        S[i], S[j] = S[j], S[i]

    return S

def generate_keystream(S, data_length):
    i = 0
    j = 0
    keystream = []

    for _ in range(data_length):
        i = (i + 1) % 256
        j = (j + S[i]) % 256
        S[i], S[j] = S[j], S[i]
        keystream_byte = S[(S[i] + S[j]) % 256]
        keystream.append(keystream_byte)

    return keystream

def rc4_encrypt(data, key):
    S = initialize_rc4(key)
    keystream = generate_keystream(S, len(data))
    encrypted_data = [byte ^ keystream_byte for byte, keystream_byte in zip(data, keystream)]

    return bytes(encrypted_data)

def rc4_decrypt(encrypted_data, key):
    return rc4_encrypt(encrypted_data, key)  # RC4 decryption is the same as encryption

# Example usage:
if __name__ == "__main__":
    plaintext = b"Hello, RC4!"
    key = b"SecretKey"

    encrypted_text = rc4_encrypt(plaintext, key)
    decrypted_text = rc4_decrypt(encrypted_text, key)

    print(f"Plaintext: {plaintext}")
    print(f"Encrypted: {encrypted_text}")
    print(f"Decrypted: {decrypted_text.decode('utf-8')}")
```
